```python
ref_arquivo = open("questoes.txt","r")

qtd = [0,0,0]
count = -1

nasResp = 0

respostas = []
perguntas = [[], [], []]

alternativasA = [[], [], []]
alternativasB = [[], [], []]
alternativasC = [[], [], []]
alternativasD = [[], [], []]


for linha in ref_arquivo:
    if linha == "### Answers\n":
        nasResp = 1
        count = 0
    else:
        if nasResp == 0:
            if linha == "### Fácil\n":                
                count = count + 1     
            elif linha == "### Médio\n":
                count = count + 1 
            elif linha == "### Difícil\n":
                count = count + 1 
            elif linha != "\n" and linha[:2] != "1 " and linha[:2] != "2 " and linha[:2] != "3 " and linha[:2] != "4 ":                
                qtd[count] += 1
                perguntas[count].append(linha.replace("\n", "", 1))
            elif linha[:2] == "1 ":
                alternativasA[count].append(linha.replace("1 ", "", 1).replace("\n", "", 1))
            elif linha[:2] == "2 ":
                alternativasB[count].append(linha.replace("2 ", "", 1).replace("\n", "", 1))
            elif linha[:2] == "3 ":
                alternativasC[count].append(linha.replace("3 ", "", 1).replace("\n", "", 1))
            elif linha[:2] == "4 ":
                alternativasD[count].append(linha.replace("4 ", "", 1).replace("\n", "", 1))
        else:
            if linha != "\n":
                respostas.append(linha.replace("\n", "", 1))
                count += 1



for i in range (0, 100):
    print(perguntas[0][i%100] + "****a) " + alternativasA[0][i%100] + "****b) " + alternativasB[0][i%100] + "****c) " + alternativasC[0][i%100] + "****d) " + alternativasD[0][i%100] + "****"  + str(int(respostas[i])-1) + "****" + "1")

for i in range (0, 40):
    print()
    
for i in range (100, 200):
    print(perguntas[1][i%100] + "****a) " + alternativasA[1][i%100] + "****b) " + alternativasB[1][i%100] + "****c) " + alternativasC[1][i%100] + "****d) " + alternativasD[1][i%100] + "****"  + str(int(respostas[i])-1) + "****" + "2")
    
for i in range (0, 40):
    print()
    
for i in range (200, 300):
    print(perguntas[2][i%100] + "****a) " + alternativasA[2][i%100] + "****b) " + alternativasB[2][i%100] + "****c) " + alternativasC[2][i%100] + "****d) " + alternativasD[2][i%100] + "****"  + str(int(respostas[i])-1) + "****" + "3")

ref_arquivo.close()
```
